## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dset = pd.read_csv('Position_Salaries.csv')
X = dset.iloc[:, 1:-1].values
y = dset.iloc[:, -1].values
```

## Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression

lin_reg1 = LinearRegression()
lin_reg1.fit(X, y)
lin_reg1_y_predict = lin_reg1.predict(X)
```

## Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X) # To convert X into a matrix of features X^1, X^2........
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
lin_reg2_y_predict = lin_reg2.predict(X_poly)
```

## Visualising the Linear Regression results

```
plt.plot(X, lin_reg1_y_predict, color='red')
plt.scatter(X, y, color='green')
plt.title("Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results

```
plt.plot(X, lin_reg2_y_predict, color='red')
plt.scatter(X, y, color='green')
plt.title("Polynomial Regression")
plt.xlabel("Position Level")
plt.xlabel("Salary")
plt.show()
```

## Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new result with Linear Regression

```
print(lin_reg1.predict([[6.5]])) # Input should be a 2-D numpy array
```

## Predicting a new result with Polynomial Regression

```
print(lin_reg2.predict(poly_reg.fit_transform([[6.5]])))
```
